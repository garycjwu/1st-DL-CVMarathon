```python
import xmltodict
import os
import urllib.request
import zipfile

try:
    os.makedirs( './data', exist_ok = True )
    res = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
    urllib.request.urlretrieve(res, "./data/example.zip")
    f = zipfile.ZipFile('./data/example.zip')
    f.extractall('./data')
except:
    print('failed to download file')
```

```python
with open('./data/64_72hr_CH.xml', encoding = 'utf-8') as fReader:
    file = dict(xmltodict.parse(fReader.read()))

locations = file['cwbopendata']['dataset']['locations']['location']
print('Number of locations:', len(locations))
```

```python
for location in locations:
    print(f"Location name: {location['locationName']}")
    
    time = location['weatherElement'][0]['time'][0]
    elementValue = time['elementValue']
    
    print(f"Time: {time['dataTime']}")
    print(f"Temperature: {elementValue['value']} {elementValue['measures']}")
```
