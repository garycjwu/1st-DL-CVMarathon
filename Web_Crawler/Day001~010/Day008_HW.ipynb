```python
import requests
from bs4 import BeautifulSoup

dcard = 'https://www.dcard.tw/f'
know = 'https://www.zhihu.com/explore'
```

```python
resp_dcard = requests.get(dcard)
resp_dcard.encoding = 'utf-8'

print(resp_dcard.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ => ', type(resp_dcard.text))
```

```python
soup_dcard = BeautifulSoup(resp_dcard.text, "html5lib")
print(soup_dcard)
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')
print("型態:", type(soup_dcard))
print("原因:", '型態會變成bs4，可以更有效率地取出所需的資料')
```

```python
resp_know = requests.get(know)
resp_know.encoding = 'utf-8'

print(resp_know.text[0:600])
```

```python
print("Response is bad. Headers are needed to do web crawl.")
print("********************************")


headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}

resp_know = requests.get(know, headers = headers)
resp_know.encoding = 'utf-8'
print("Response after adding headers.")
print(resp_know.text[0:600])
```
