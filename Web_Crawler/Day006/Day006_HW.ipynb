```python
import requests
import json

url = 'https://www.zhihu.com/api/v4/questions/55493026/answers'
headers = {'user-agent' : 'my-app/0.0.1'}
r = requests.get(url, headers=headers)
response = r.text
response
```

```python
data = json.loads(response)
data
```

```python
from datetime import datetime

time_sec = data['data'][0]['question']['created']
time = datetime.fromtimestamp(time_sec)
print(f'Question created time: {time}')
```

```python
first_answer_time = data['data'][0]['created_time']
last_answer_time = data['data'][0]['created_time']

for row in data['data']:
    time = row['created_time']
    if time > last_answer_time:
        last_answer_time = time
    if time < first_answer_time:
        first_answer_time = time
    
first_time = datetime.fromtimestamp(first_answer_time)
last_time = datetime.fromtimestamp(last_answer_time)

print(f'First answer time: {first_time}')
print(f'Last answer time: {last_time}')
```
